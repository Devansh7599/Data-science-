```python
# data interpolition

#liner interpolaton
#cubic
# polinomial




#linear interpolitaion ]
import matplotlib.pyplot as plt
import numpy as np


x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

```

```python
plt.scatter(x,y)
```

```python
x_new=np.linspace(1,5,10)
y_interp=np.interp(x_new,x,y)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)

```

```python
# cubic interpolation
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f=interp1d(x,y,kind='cubic')
```

```python
x_new=np.linspace(1,5,10)
y_interp=f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

```python
#poltinomial interpoltion

x=np.array([1,2,3,4,5])
y=np.array([1,4,9,1,25])
```

```python
# interpolate using polinomial interpolation

p=np.polyfit(x,y,2)
```

```python
x_new=np.linspace(1,5,10)
y_interp=np.polyval(p,x_new)
```

```python
y_interp
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sallary=[11,40,45,68,65,68,78,90,57,74,91,92,88,68,57,48,99]
```

```python
# five point summery
#min  25th,50th,75th,100th

np.quantile(sallary,[0,0.25,0.5,0.75,1])
```

```python
df=pd.Series(sallary).describe()
```

```python
# to check outliers
#displot, boxplot

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.histplot( sallary, kde =True)
plt.title('hist plot')
plt.subplot(1,2,2)
sns.boxplot( x = sallary)
plt.title('box plot')
```
